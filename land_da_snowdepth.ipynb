```python
import matplotlib.pyplot as plt
import netCDF4
```

```python
# open a local NetCDF file at /lustre/LandDA/landda_expts/DA_GHCN_test/DA/jedi_incr
file = '20160102.180000.xainc.sfc_data.tile3.nc'
data = netCDF4.Dataset(file)

# Extract the snwdph data
snwdph = data.variables['snwdph'][0, 0, :, :]
print(data.variables.keys())

# Extract xaxis_1 and yaxis_1 variables
xaxis_1 = data.variables['xaxis_1'][:]
yaxis_1 = data.variables['yaxis_1'][:]

# Close the NetCDF file
data.close()

plt.imshow(snwdph, cmap='jet', extent=[xaxis_1[0], xaxis_1[-1], yaxis_1[0], yaxis_1[-1]])
plt.colorbar(label='Snow Depth (cm)')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title('SNOW DEPTH')

# Save the plot as an image
plt.savefig('snwdph_image.png', dpi=300)

# Show the plot
plt.show()
```
